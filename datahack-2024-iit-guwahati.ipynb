```python
numerical_cols = X.select_dtypes(include = ["int", "float"]).columns
categorical_cols = X.select_dtypes(include = ["object"]).columns
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state = 0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

numerical_transformer = SimpleImputer(strategy='constant')

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.neural_network import MLPClassifier
import xgboost
from xgboost import XGBClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import roc_auc_score

model_1 = MultiOutputClassifier(LogisticRegression(max_iter=500), n_jobs=-1)
model_2 = MultiOutputClassifier(RandomForestClassifier(), n_jobs=-1)
model_3 = MultiOutputClassifier(GradientBoostingClassifier(), n_jobs=-1)
model_4 = MultiOutputClassifier(XGBClassifier(), n_jobs=-1)
model_5 = MultiOutputClassifier(MLPClassifier(max_iter=500), n_jobs=-1)
```

```python
def check_model(preprocessor, model, X_train, y_train, X_valid, y_valid):
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)
                             ])
    pipeline.fit(X_train, y_train)
    predictions_prob = pipeline.predict_proba(X_valid)
    predictions = np.array([predictions_prob[0][:, 1], predictions_prob[1][:, 1]])
    predictions = predictions.T
    print(roc_auc_score(y_valid, predictions))
```

```python
check_model(preprocessor, model_1, X_train, y_train, X_valid, y_valid)
```

```python
check_model(preprocessor, model_2, X_train, y_train, X_valid, y_valid)
```

```python
check_model(preprocessor, model_3, X_train, y_train, X_valid, y_valid)
```

```python
check_model(preprocessor, model_4, X_train, y_train, X_valid, y_valid)
```

```python
check_model(preprocessor, model_5, X_train, y_train, X_valid, y_valid)
```
